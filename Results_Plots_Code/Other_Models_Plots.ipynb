```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import os
ls_dnn_csv = os.listdir(dnn_path_csv)
ls_others_csv = os.listdir(other_models_path_csv)
```

```python
# For dijkstra mape only from 2nd iterations to last
def plot_boxplot_new_2(df, file, val):
    ls = ['dnn_1', 'dnn_2', 'dnn_3', 'dnn_4']
    new_df = pd.DataFrame(columns = ls) 
    for i in range(len(ls)):
        arr = df['mape'][i][1:-1].split(',') 
        temp = []
        j = 0
        for ele in arr:
            if j%2 == 0:
                temp.append(ele)
            j = j + 1    
        # print(temp)
        arr_new = [float(s[8:-2]) for s in temp]
        new_df[ls[i]] = arr_new
    # print(new_df.head())
    plt.figure()
    boxplot = new_df.boxplot(column = ls)
    plt.title(file)
```

```python
# For dijkstra mape only for 1st iteration
def plot_boxplot_new(df, file, val):
    ls = ['dnn_1', 'dnn_2', 'dnn_3', 'dnn_4']
    new_df = pd.DataFrame(columns = ls) 
    for i in range(len(ls)):
        arr = [float(s[8:-2]) for s in df[val][i][1:-1].split(',')]
        new_df[ls[i]] = arr
    # print(new_df.head())
    plt.figure()
    boxplot = new_df.boxplot(column = ls)
    plt.title(file)
```

```python
def plot_boxplot(df, file, val):
    ls = ['best_svr', 'best_lr', 'best_rr', 'best_knn', 'best_gpr','best_dt','best_rf','best_etr','best_gbr', 'best_xgb']
    new_df = pd.DataFrame(columns = ls) 
    for i in range(len(ls)):
        arr = [float(s) for s in df[val][i][1:-1].split(',')]
        new_df[ls[i]] = arr
    # print(new_df.head())
    plt.figure()
    boxplot = new_df.boxplot(column = ls, rot = 90)
    plt.title(file)
```

# Plotting R2 scores of 10 models other than DNN

```python
for file in ls_others_csv:
    df = pd.read_csv(other_models_path_csv + '/' + file)
    plot_boxplot(df, file,val = 'r2')
```

```python
for file in ls_others_csv:
    df = pd.read_csv(other_models_path_csv + '/' + file)
    plot_boxplot(df, file,val = 'mse')
```
