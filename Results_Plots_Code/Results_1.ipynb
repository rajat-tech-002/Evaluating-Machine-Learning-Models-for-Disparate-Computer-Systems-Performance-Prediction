```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import pyplot
import numpy as np
plt.rcParams.update({'figure.max_open_warning': 0})
```

```python
df = pd.read_csv('Results_1 - Grid Search without and with.csv')
```

```python
df.head()
```

```python
ls  = df['Algo']
ls
```

```python
df.iloc[0]['R2']
```

```python
def convert_to_num(arr, ind):
    # print(arr[ind])
    arr = arr[ind]
    arr = [float(s) for s in arr[1:-1].split(',')]
    # print(arr)
    return arr
```

```python
def convert_to_num_mape(arr, ind):
    # print(arr[ind])
    arr = arr[ind]
    arr = [float(s) for s in arr[1:-1].split(',')]
    arr = [x*100 for x in arr]
    # print(arr)
    return arr
```

```python
x = ls
y1 = []
for i in range(8):
    # x = [ls[0], ls[1]]
    a = convert_to_num(df[df['Algo'] ==ls[i]]['R2'], ind = i)
    # b = convert_to_num(df[df['Algo'] == ls[1]]['R2'], ind = 1)
    # y = [a, b ]
    y1.append(a)
plt.boxplot(y1, labels = x)
plt.xlabel('Models with and without GS')
plt.ylabel('R2 score')
plt.title('Comparison of R2 scores')
plt.xticks(rotation = 45)
# plt.savefig('Results_1.jpg')
```

```python
x = ls
y2 = []
for i in range(8):
    # x = [ls[0], ls[1]]
    a = convert_to_num_mape(df[df['Algo'] ==ls[i]]['MedAPE'], ind = i)
    # b = convert_to_num(df[df['Algo'] == ls[1]]['R2'], ind = 1)
    # y = [a, b ]
    y2.append(a)
plt.boxplot(y2, labels = x)
plt.xlabel('Models with and without GS')
plt.ylabel('MedAPE score')
plt.title('Comparison of MedAPE scores')
plt.xticks(rotation = 45)
# plt.savefig('Results_1.jpg')
```

```python
x = ls
y3 = []
for i in range(8):
    # x = [ls[0], ls[1]]
    a = convert_to_num(df[df['Algo'] ==ls[i]]['MSE'], ind = i)
    # b = convert_to_num(df[df['Algo'] == ls[1]]['R2'], ind = 1)
    # y = [a, b ]
    y3.append(a)
plt.boxplot(y3, labels = x)
plt.xlabel('Models with and without GS')
plt.ylabel('MSE score')
plt.title('Comparison of MSE scores')
plt.xticks(rotation = 45)
# plt.savefig('Results_1.jpg')
```
